```python
import yfinance as yf
```

```python
sp500= yf.Ticker("^GSPC")
```

```python
sp500=sp500.history(period="max")
```

```python
 sp500.index
```

```python
#Cleaning and visualizing our stock market Data
sp500.plot.line(y="Close",use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#setting up our Target for Machine Learning(It tells whether stock goes up or down)
# Creating a one extra column that is "Tommorow"
sp500["Tommorow"]=sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"]=(sp500["Tommorow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
#Too much of historocal Data is harmfarmfull in terms of stock market so we are deleting before 1990 data.
#.copy.() used bcoz sometimes pandas gives setting warning especially when we are assining subset to a same set 
sp500=sp500.loc["1990-01-01":].copy()
```

```python
sp500
```

```python
#Traing An Initial Machine learning model
#Reasons for using RandomForestClassifier is 
#1.RandomForestClassifier works by training a bunch of individual decision trees with randomized parameters.
#2.And then averaging the Results from those decision trees 
#3.Bcoz of these process Randomforest resistant to overfitting they can but its harder them to overfit than other models 
#4.it can find linear relationships more accurate

from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=1)
train=sp500.iloc[:-100]
test=sp500.iloc[-100:]

predictors=["Close","Volume","Open","High","Low"]
model.fit(train[predictors],train["Target"])
```

```python
#Measure how accurate the model using metrics precision_score
from sklearn.metrics import precision_score

preds=model.predict(test[predictors])
```

```python
import pandas as pd

preds=pd.Series(preds,index=test.index)
```

```python
precision_score(test["Target"],preds)
```

```python
#Improving the model
combined=pd.concat([test["Target"],preds],axis=1)
```

```python
combined.plot()
```

```python
#Building a Backtesting system
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict(test[predictors])
    preds=pd.Series(preds,index=test.index,name="Predictions")
    combined=pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
def backtest(data,model,predictors,start=2500,step=250):
    all_predictions=[]
    
    for i in range(start,data.shape[0],step):
        train=data.iloc[0:i].copy()
        test=data.iloc[i:(i+step)].copy()
        predictions= predict(train,test,predictors,model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions=backtest(sp500,model,predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"],predictions["Predictions"])
```

```python
#Adding additional predictors in our model
horizons=[2,5,60,250,1000]
new_predictors = []
for horizon in horizons:
    rolling_averages=sp500.rolling(horizon).mean()
    
    ratio_column=f"Close_Ratio_{horizon}"
    sp500[ratio_column]=sp500["Close"]/rolling_averages["Close"]
    
    
    trend_column=f"Trend_{horizon}"
    sp500[trend_column]=sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+=[ratio_column,trend_column]
```

```python
sp500=sp500.dropna()
sp500.head()
```

```python
#Improving our model
model=RandomForestClassifier(n_estimators=200,min_samples_split=50,random_state=1)
```

```python
def predict(train,test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict_proba(test[predictors])[:,1]# gives probability 
    preds[preds>=0.6]=1 #These reduce total no.of trading days but it will increase the chance that price actually go up
    preds[preds<0.6]=0
    preds=pd.Series(preds,index=test.index,name="Predictions")
    combined=pd.concat([test["Target"],preds],axis=1)
    return combined
```

```python
predictions=backtest(sp500,model,new_predictors)
```

```python
#If we observe no.of positive trading days is reduced.
predictions["Predictions"].value_counts()
```

```python
#According to Time series data its good to have more than probability of 50% Success Rate
precision_score(predictions["Target"],predictions["Predictions"])
```
